```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import data
df = pd.read_csv('../Final Data/Final_Data.csv')
df.head(5)
```

```python
df.describe()
```

```python
df_corr = df.drop(columns = 'Date')
df_corr.corr()
```

```python
sns.pairplot(df_corr, diag_kind='kde')
plt.suptitle('Pair Plot of Wheat Data', y=1.02)
plt.show()
```

```python
# Plot time series data
plt.figure(figsize=(14, 8))
plt.plot(df['Date'], df['Wheat_Price (Dollars per Bushel)'], 'r-', linewidth=1.4)

plt.xlabel('Date')
plt.ylabel('Wheat Price ($/Bushel)')

plt.title('../Visualizations/Wheat and Flour Prices Over Time')
plt.tight_layout()
plt.show()
```

```python
#Plot Price Change by year
df['Price_Change'] = df['Wheat_Price (Dollars per Bushel)'].diff()
plt.figure(figsize=(12, 8))
plt.plot(df['Date'], df['Price_Change'], 'b-', linewidth=1.2)
plt.title('Wheat Price Change Over Time')
plt.ylabel('Wheat Price Change (USD)')
plt.xlabel('Year')
plt.axhline(y=0, color='g', linestyle='--')
plt.tight_layout()
plt.savefig('../Visualizations/Change_in_Wheat_Price_Over_Time.png')
plt.show()
```

```python
#plot of absolute value of price change
df['Price_Change_abs'] = df['Price_Change'].abs()
plt.figure(figsize=(12, 8))
plt.plot(df['Date'], df['Price_Change_abs'], 'b-')
plt.title('Absolute Value of Wheat Price Change Over Time')
plt.ylabel('Absolute Wheat Price Change')
plt.xlabel('Year')
plt.tight_layout()
plt.savefig('../Visualizations/Absolute_Change_Over_Time.png')
plt.show()
```
