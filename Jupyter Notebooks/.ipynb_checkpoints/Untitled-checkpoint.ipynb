```python
#import libraries
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
from statsmodels.tools.sm_exceptions import ConvergenceWarning
warnings.filterwarnings('ignore', category=ConvergenceWarning)
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.ar_model import AutoReg
from statsmodels.tsa.ar_model import ar_select_order
import itertools
from statsmodels.graphics.tsaplots import plot_acf
import statsmodels.api as sm
from scipy import stats
import matplotlib.pyplot as plt
from statsmodels.tsa.api import VAR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, HistGradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from arch import arch_model
import seaborn as sns
from sklearn.model_selection import GridSearchCV
from sklearn.inspection import permutation_importance
from datetime import datetime
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, GRU, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
import xgboost as xgb
from sklearn.model_selection import GridSearchCV, TimeSeriesSplit
```

```python
df = pd.read_csv('../Final Data/Final_Data.csv')
df['date'] = pd.to_datetime(df['date'])

df.head()
```

```python
exog_vars = ['fertilizer_idx', 'avg_temp', 'drought_idx', 
             'cpi', 'commodity_imp_idx', 'commodity_exp_idx',
             'flour_price', 'corn_price', 'crude_oil_price']

for var in exog_vars:
    for lag in range(1, 7):
        df[f'{var}_lag{lag}'] = df[var].shift(lag)

lagged_features = [col for col in df.columns if 'lag' in col]

scaler = StandardScaler() #standardize
df[lagged_features] = scaler.fit_transform(df[lagged_features])

#change date to month and year vars
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month

for lag in range(1, 7):
        df[f'wheat_price_lag{lag}'] = df['wheat_price'].shift(lag)
df = df.dropna()

#drop non-lagged features and date
df = df.drop(columns = ['fertilizer_idx', 'avg_temp','drought_idx','cpi',
                        'commodity_imp_idx','commodity_exp_idx','flour_price','corn_price',
                       'crude_oil_price','date']).reset_index(drop=True)

df.head()
```

```python
#split to train/test
y = df['wheat_price']
X = df.drop(columns='wheat_price')
train_size = int(len(df) * 0.8)
train = df.iloc[:train_size]
test = df.iloc[train_size:]
```

```python
results = {}
predictions_df = pd.DataFrame({'Actual': test['wheat_price']})
```

```python
ar_select = ar_select_order(train['wheat_price'], maxlag=12)
optimal_lag = ar_select.ar_lags

print(f"Optimal AR lag: {optimal_lag}")
```

```python
ar1_model = AutoReg(train['wheat_price'], lags=1)
ar1_results = ar1_model.fit()
print('AR(1) Results: \n',ar1_results.summary())
ar2_model = AutoReg(train['wheat_price'], lags=2)
ar2_results = ar2_model.fit()
print('\nAR(2) Results: \n',ar2_results.summary())
```

```python
#Forecast
ar1_preds = ar1_results.predict(start=len(train), end=len(train) + len(test) - 1)
ar1_mse = mean_squared_error(test['wheat_price'], ar1_preds)
results['AR1'] = ar1_mse
predictions_df['AR1'] = ar1_preds

ar2_preds = ar2_results.predict(start=len(train), end=len(train) + len(test) - 1)
ar2_mse = mean_squared_error(test['wheat_price'], ar2_preds)
results['AR2'] = ar2_mse
predictions_df['AR2'] = ar2_preds
```

```python
best_aic = float('inf')
best_order = None

for p in range(3):  # AR order
    for d in range(2):  # Differencing
        for q in range(3):  # MA order
            try:
                model = ARIMA(train['wheat_price'], order=(p, d, q))
                model_fit = model.fit()
                if model_fit.aic < best_aic:
                    best_aic = model_fit.aic
                    best_order = (p, d, q)
            except:
                continue

print(f"Best ARIMA order: {best_order}")

# Fit ARIMA with best order
arima_model = ARIMA(train['wheat_price'], order=best_order)
arima_fit = arima_model.fit()
print(arima_fit.summary())
```

```python
# Forecast
arima_preds = arima_fit.forecast(steps=len(test))
arima_mse = mean_squared_error(test['wheat_price'], arima_preds)
results['ARIMA'] = arima_mse
predictions_df['ARIMA'] = arima_preds
```

```python
exog_lagged_vars = [col for col in X.columns if 'lag' in col and 'wheat_price' not in col]
train_exog = train[exog_lagged_vars]
test_exog = test[exog_lagged_vars]
print(train_exog.head())
```

```python
best_aic = float('inf')
best_order = None

for p in range(3):  # AR order
    for d in range(2):  # Differencing
        for q in range(3):  # MA order
            try:
                arimax_model = SARIMAX(train['wheat_price'], exog=train_exog,order=(p,d,q),seasonal_order=(0, 0, 0, 0))
                model_fit = arimax_model.fit(disp=False)
                if model_fit.aic < best_aic:
                    best_aic = model_fit.aic
                    best_order = (p, d, q)
            except:
                continue

print(f"Best ARIMAX order: {best_order}")

arimax_fit = arimax_model.fit(disp=False)
print(arimax_fit.summary())
```

```python
# Forecast
arimax_preds = arimax_fit.forecast(steps=len(test), exog=test_exog)
arimax_mse = mean_squared_error(test['wheat_price'], arimax_preds)
results['ARIMAX'] = arimax_mse
predictions_df['ARIMAX'] = arimax_preds
```

```python
# Check for seasonality patterns with period=12 (monthly)
best_seasonal_aic = float('inf')
best_order = None
best_seasonal_order = None
best_model = None

# Keep track of progress
total_models = 3*2*3*2*2*2
current = 0

for p in range(3):  # AR order
    for d in range(2):  # Differencing
        for q in range(3):
            for P in range(2):
                for D in range(2):
                    for Q in range(2):
                        current += 1
                        print(f"Testing model {current}/{total_models}: SARIMAX({p},{d},{q})({P},{D},{Q},12)", end="\r")
                        
                        try:
                            model = SARIMAX(
                                train['wheat_price'], 
                                exog=train_exog,
                                order=(p,d,q),
                                seasonal_order=(P, D, Q, 12)
                            )
                            model_fit = model.fit(disp=False, maxiter=50)  # Limit iterations for initial search
                            
                            if model_fit.aic < best_seasonal_aic:
                                best_seasonal_aic = model_fit.aic
                                best_order = (p,d,q)
                                best_seasonal_order = (P, D, Q, 12)
                                best_model = model_fit
                                print(f"\nNew best model: SARIMAX{best_order}{best_seasonal_order} AIC: {best_seasonal_aic:.2f}")
                                
                        except Exception as e:
                            # Print first 50 chars of the error
                            print(f"\nError with SARIMAX({p},{d},{q})({P},{D},{Q},12): {str(e)[:50]}...")
                            continue

print(f"\nBest SARIMAX order: {best_order}")
print(f"Best seasonal order: {best_seasonal_order}")
print(f"Best AIC: {best_seasonal_aic:.2f}")

# Use saved best model rather than refitting
if best_model is not None:
    print(best_model.summary())
```

```python
#Forecast
sarimax_preds = model_fit.forecast(steps=len(test), exog=test_exog)
sarimax_mse = mean_squared_error(test['wheat_price'], sarimax_preds)
results['SARIMAX'] = sarimax_mse
predictions_df['SARIMAX'] = sarimax_preds
```

```python
X_train = train.drop(columns='wheat_price')
y_train = train['wheat_price']
X_test = test.drop(columns='wheat_price')
y_test = test['wheat_price']
print(train.head())
```

```python
params = {
    'max_depth': [3, 5, 7, 10, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

#run grid search cross-validation for tuning
dt_model = DecisionTreeRegressor(random_state=42) #initialize
dt_gridsearch = GridSearchCV(dt_model, params,cv=5) #start gridsearch
dt_gridsearch.fit(X_train, y_train) #fit model
best_params = dt_gridsearch.best_params_ #find best parameters
print(f"Best Parameters: {best_params}")

best_dt_model = DecisionTreeRegressor(**best_params, random_state=42)
best_dt_model.fit(X_train, y_train)
```

```python
dt_preds = best_dt_model.predict(X_test)
dt_mse = mean_squared_error(y_test, dt_preds)
results['Decision Tree'] = dt_mse
predictions_df['Decision Tree'] = dt_preds
```

```python
tscv = TimeSeriesSplit(n_splits=5)
xgb_model = xgb.XGBRegressor(
    objective='reg:squarederror',  # For regression problems
    random_state=42               # For reproducibility
)
# Define parameter grid to search
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.05, 0.1],
    'subsample': [0.8, 0.9, 1.0],
    'colsample_bytree': [0.8, 0.9, 1.0],
    'min_child_weight': [1, 3, 5],
    'gamma': [0, 0.1, 0.2]
}

grid_search = GridSearchCV(
    estimator=xgb_model,
    param_grid=param_grid,
    scoring='neg_mean_squared_error',  # Use negative MSE as the scoring metric
    cv=tscv,                           # Time series cross-validation
    verbose=1,
    n_jobs=-1                          # Use all available cores
)

# Perform grid search
grid_search.fit(X_train, y_train)

# Print best parameters and score
print("Best parameters found:")
print(grid_search.best_params_)
print(f"Best cross-validation score: {-grid_search.best_score_}")

best_xgb_model = xgb.XGBRegressor(**grid_search.best_params_, random_state=42)
best_xgb_model.fit(X_train, y_train)

# Evaluate on test set
best_xgb_preds = best_xgb_model.predict(X_test)
best_xgb_mse = mean_squared_error(y_test, best_xgb_preds)
results['XGBoost (Tuned)'] = best_xgb_mse
predictions_df['XGBoost (Tuned)'] = best_xgb_preds
```

```python
tuned_importance = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': best_xgb_model.feature_importances_
}).sort_values('Importance', ascending=False)

print("Tuned XGBoost Feature Importance:")
print(tuned_importance.head(10))
```

```python
results_df = pd.DataFrame({
    'Model': results.keys(),
    'MSE': results.values()
}).sort_values('MSE')

print("\nModel Comparison Results:")
print("-" * 40)
print(results_df)
```
